## Part 1
Create a plot of $y = \sin(x)$ from 0 to 2$\pi$.  Remember to use Numpy for math functions and add useful comments to your code.

```python
import numpy as np
import matplotlib.pyplot as plt

X = np.linspace(0, 2*np.pi)

plt.ylim(-1.5, 1.5)
plt.plot(X, np.sin(X))
plt.ylabel('y')
plt.xlabel('x')
plt.show()
```
